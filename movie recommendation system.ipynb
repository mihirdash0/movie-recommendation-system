```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=10000)
tfidf_matrix = tfidf.fit_transform(new_df['tag'])
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=100)  # adjust to balance speed/accuracy
reduced_matrix = svd.fit_transform(tfidf_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn_model = NearestNeighbors(n_neighbors=6, metric='cosine').fit(reduced_matrix)
```

```python
indices = pd.Series(new_df.index, index=new_df['title']).drop_duplicates()
```

```python
from sklearn.cluster import KMeans

# Try 100 clusters (tune based on dataset)
num_clusters = 90
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
new_df['cluster'] = kmeans.fit_predict(reduced_matrix)
```

```python
def recommend_cluster(title, n=10):
    if title not in df['title'].values:
        return ["Movie not found."]
    
    # Find cluster of given movie
    movie_cluster = new_df.loc[new_df['title'] == title, 'cluster'].values[0]
    
    # Get other movies from same cluster
    cluster_movies = new_df[new_df['cluster'] == movie_cluster]
    
    # Exclude the input movie
    recs = cluster_movies[cluster_movies['title'] != title]
    
    # Return top N based on similarity or just random
    return recs['title'].head(n).tolist()

```
